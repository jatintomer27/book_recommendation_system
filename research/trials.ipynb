```python
ratings_with_books.head()
```

```python
# It group the data based on title and select only the ratings column
# And count how many number of times each book has rated

ratings_with_books.groupby("title")['rating'].count()
```

```python
# Here reset_index() convert the series back into the dataframe

number_ratings = ratings_with_books.groupby("title")['rating'].count().reset_index()
number_ratings.head()
```

```python
final_ratings = ratings_with_books.merge(number_ratings, on="title")
final_ratings.head()
```

```python
final_ratings.columns
```

```python
final_ratings.rename(columns={
    'rating_y':'ratings_count',
    'rating_x':'rating'
},inplace=True)
```

```python
final_ratings.columns
```

```python
final_ratings.head()
```

```python
# Filter those books who has got atleast 50 ratings


final_ratings = final_ratings[final_ratings["ratings_count"] > 50]
final_ratings.head()
```

```python
final_ratings.shape
```

```python
# Drop the duplicate rows

final_ratings.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_ratings.shape
```

```python
# Check for Null values

final_ratings.isnull().sum()
```

```python
# Create the pivot table from the data

book_pivot = final_ratings.pivot_table(columns='user_id',index='title', values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot
```

```python
# I can't pass NaN data to the model therefore we have to convert it into the 0

book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
# Convert this matrix to the CSR Matrix

from scipy.sparse import csr_matrix

book_spare = csr_matrix(book_pivot)
```

```python
book_spare
```

```python
type(book_spare)
```

### What is CSR Matrix and why it is used**

- A CSR matrix means Compressed Sparse Row matrix.

- Itâs a special way to store a sparse matrix (a matrix where most values are 0), so we donât waste memory storing all those zeros.

![CSR 1](images/csr_1.png)
![CSR 2](images/csr_2.png)
![CSR 3](images/csr_3.png)

----------

```python
# Initilize the model

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
# Train the model

model.fit(book_spare)
```

```python
# Test whether the trained model is giving the recommendation or not

# I will give a book to the model and it will give me books 
# that i can suggest based on the trained model

book_pivot.iloc[237,:]
```

```python
book_pivot.iloc[237,:].values
```

```python
book_pivot.iloc[237,:].values.shape
```

```python
test_data = book_pivot.iloc[237,:].values.reshape(1,-1)
```

```python
test_data.shape
```

```python
# Predict the 6 books based on the trained model

distance, suggestions = model.kneighbors(test_data, n_neighbors=6)
```

```python
# Distance with 0 is the same book we are passing to predict

distance
```

## Final recommendation function 

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    for suggestion in suggestions[0]:
        books = book_pivot.index[suggestion]
        print(books)
```

```python
book_name = "2nd Chance"
recommend_book(book_name)
```
